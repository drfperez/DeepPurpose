```python

!pip install rdkit
!pip install mordred

from multiprocessing import freeze_support
from rdkit import Chem
from mordred import Calculator, descriptors
from google.colab import files
import io
import pandas as pd

def generate_descriptors_from_smiles(smiles_list):
    # Convert SMILES strings to RDKit molecules
    mols = [Chem.MolFromSmiles(smiles) for smiles in smiles_list if Chem.MolFromSmiles(smiles) is not None]

    if len(mols) == 0:
        print("No valid molecules found in the input SMILES list.")
        return None

    # Create Calculator
    calc = Calculator(descriptors)

    # Calculate descriptors
    descriptors_result = calc.pandas(mols)

    return descriptors_result

def download_descriptors_csv(descriptors_df):
    # Save dataframe to a string buffer
    buffer = io.StringIO()
    descriptors_df.to_csv(buffer, index=False)
    buffer.seek(0)

    # Download the CSV file
    files.download("molecular_descriptors.csv")

if __name__ == "__main__":
    freeze_support()

    # Accept user input
    smiles_input = input("Enter SMILES strings separated by commas: ")
    smiles_list = smiles_input.split(",")

    # Generate descriptors
    descriptors_df = generate_descriptors_from_smiles(smiles_list)

    if descriptors_df is not None:
        # Download the resulting dataframe as a CSV file
        download_descriptors_csv(descriptors_df)
```
