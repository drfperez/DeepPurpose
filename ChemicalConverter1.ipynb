```python
# Convert mol or sdf files to mol2 file

from google.colab import files
from openbabel import openbabel as ob

# Upload the file
uploaded = files.upload()

# Get the uploaded file name
file_name = list(uploaded.keys())[0]

# Check if file is uploaded
if not file_name:
    print("Error: No file uploaded.")
else:
    # Initialize the Open Babel molecule
    mol = ob.OBMol()

    # Read the file
    conv = ob.OBConversion()

    # Set input format based on file extension
    if file_name.endswith('.sdf'):
        conv.SetInFormat("sdf")
    elif file_name.endswith('.mol'):
        conv.SetInFormat("mol")
    else:
        print("Error: Unsupported file format.")

    if conv.ReadFile(mol, file_name):
        # Add all hydrogens
        mol.AddHydrogens()

        # Write the mol2 file
        output_file_name = file_name.split('.')[0] + '_with_hydrogens.mol2'
        conv.SetOutFormat("mol2")
        conv.WriteFile(mol, output_file_name)

        # Save the mol2 file locally
        with open(output_file_name, 'w') as f:
            f.write(conv.WriteString(mol))

        print("Mol2 file with added hydrogens generated successfully!")
        # Download the mol2 file
        files.download(output_file_name)
    else:
        print("Error: Unable to read the file.")
```

```python
# CHECK MOL2
!pip install rdkit
import os
from rdkit import Chem
from google.colab import files

# Upload a mol2 file
uploaded = files.upload()

# Get the uploaded file
mol2_file = list(uploaded.keys())[0]

# Read the mol2 file
mol = Chem.MolFromMol2Block(uploaded[mol2_file].decode("utf-8"))

# Check for errors
if mol is None:
    print("Error: Unable to read mol2 file")
else:
    print("Mol2 file read successfully")

# Check atom and bond counts
print(f"Number of atoms: {mol.GetNumAtoms()}")
print(f"Number of bonds: {mol.GetNumBonds()}")

# Check atom types and charges
for atom in mol.GetAtoms():
    print(f"Atom {atom.GetIdx()}: {atom.GetSymbol()} with charge {atom.GetFormalCharge()}")

# Check bond orders and aromaticity
for bond in mol.GetBonds():
    print(f"Bond {bond.GetIdx()}: {bond.GetBondType()} ({bond.GetIsAromatic()})")

# Check molecular weight and formula
print(f"Molecular weight: {Chem.Descriptors.MolWt(mol):.2f}")
print(f"Molecular formula: {Chem.rdmolfiles.MolToSmiles(mol)}")

# Check stereochemistry
for atom in mol.GetAtoms():
    if atom.GetChiralTag() != Chem.rdchem.ChiralType.UNDEFINED:
        print(f"Atom {atom.GetIdx()} is chiral")
```

```python

# InstalÂ·lar les llibreries necessÃ ries
!pip install rdkit py3Dmol

# Importar les llibreries necessÃ ries
from rdkit import Chem
from rdkit.Chem import AllChem
import py3Dmol
from google.colab import files

# Carregar el fitxer de l'usuari
uploaded = files.upload()

# Llegir el fitxer pujat i convertir-lo a una llista de molÃ¨cules
mol_list = []
for filename in uploaded.keys():
    with open(filename, 'rb') as f:
        mol_block = f.read().decode('utf-8')
    # RDKit no pot llegir arxius SDF amb diverses molÃ¨cules, de manera que iterem sobre elles i les afegim a la llista
    if filename.endswith('.sdf'):
        suppl = Chem.SDMolSupplier()
        suppl.SetData(mol_block)
        for mol in suppl:
            if mol is not None:
                mol_list.append(mol)
    else:
        mol = Chem.MolFromMol2Block(mol_block)
        if mol is not None:
            mol_list.append(mol)

# Visualitzar les molÃ¨cules
viewer = py3Dmol.view(width=800, height=600)
for mol in mol_list:
    pdb_block = Chem.MolToPDBBlock(mol)
    viewer.addModel(pdb_block, 'pdb')

viewer.setStyle({'stick': {}})
viewer.zoomTo()
viewer.show()
```
