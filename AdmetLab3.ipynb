```python

import requests
import pandas as pd
from io import StringIO
from google.colab import files

baseUrl = 'https://admetlab3.scbdd.com'

def save_as_json(json_response):
    with open('result.json', 'w') as f:
        f.write(json_response)
    print("JSON file saved successfully!")

if __name__ == '__main__':
    api = '/api/admet'
    url = baseUrl + api

    # Take input from the user for SMILES separated by commas
    smiles_input = input("Enter SMILES separated by commas: ")
    smiles_list = smiles_input.split(',')

    param = {
        'SMILES': smiles_list
    }

    response = requests.post(url, json=param)

    if response.status_code == 200:
        json_response = response.text
        print(json_response)
        save_as_json(json_response)
        files.download('result.json')  # Download the JSON file
    else:
        print("Failed to retrieve data from the API.")
```
