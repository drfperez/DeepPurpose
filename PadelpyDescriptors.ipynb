```python

from padelpy import from_smiles
from google.colab import files

# Prompt the user to enter SMILES strings
smiles_input = input("Enter SMILES strings separated by commas: ")

# Split the input into a list of SMILES strings
smiles_list = [smiles.strip() for smiles in smiles_input.split(',')]

# Calculate molecular descriptors for each SMILES string
descriptors_list = []
for smiles in smiles_list:
    descriptors = from_smiles(smiles)
    descriptors_list.append(descriptors)

# Save descriptors to a CSV file
from_smiles(smiles_list, output_csv='descriptors.csv')

# Download the file to your local device
files.download('descriptors.csv')
```

```python

from padelpy import from_smiles, from_mdl, from_sdf, padeldescriptor

def get_descriptors_from_smiles():
    smiles = input("Enter the SMILES string(s) separated by comma: ").split(',')
    descriptors = from_smiles(smiles)
    print(descriptors)

def get_descriptors_from_mdl():
    mdl_file = input("Enter the path to the MDL MolFile: ")
    descriptors = from_mdl(mdl_file)
    print(descriptors)

def get_descriptors_from_sdf():
    sdf_file = input("Enter the path to the SDF file: ")
    descriptors = from_sdf(sdf_file)
    print(descriptors)

def command_line_wrapper():
    print("Enter your desired parameters for PaDEL-Descriptor:")
    parameters = input("Parameters: ")
    padeldescriptor(parameters)

def main():
    while True:
        print("\nWelcome to PaDELPy Descriptor Calculator")
        print("Choose the type of input:")
        print("1. SMILES string")
        print("2. MDL MolFile")
        print("3. SDF file")
        print("4. Command Line Wrapper")
        print("5. Exit")

        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == '1':
            get_descriptors_from_smiles()
        elif choice == '2':
            get_descriptors_from_mdl()
        elif choice == '3':
            get_descriptors_from_sdf()
        elif choice == '4':
            command_line_wrapper()
        elif choice == '5':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```

```python

import csv
import os
from padelpy import from_smiles, from_mdl, from_sdf, padeldescriptor

def get_descriptors_from_smiles():
    smiles = input("Enter the SMILES string(s) separated by comma: ").split(',')
    descriptors = from_smiles(smiles)
    save_to_csv(descriptors)

def get_descriptors_from_mdl():
    mdl_file = input("Enter the path to the MDL MolFile: ")
    descriptors = from_mdl(mdl_file)
    save_to_csv(descriptors)

def get_descriptors_from_sdf():
    sdf_file = input("Enter the path to the SDF file: ")
    descriptors = from_sdf(sdf_file)
    save_to_csv(descriptors)

def command_line_wrapper():
    print("Enter your desired parameters for PaDEL-Descriptor:")
    parameters = input("Parameters: ")
    padeldescriptor(parameters)

def save_to_csv(descriptors):
    output_csv = input("Enter the path to save the CSV file: ")
    with open(output_csv, 'w', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=descriptors[0].keys())
        writer.writeheader()
        for row in descriptors:
            writer.writerow(row)
    print(f"Descriptors saved to {output_csv}")

def main():
    while True:
        print("\nWelcome to PaDELPy Descriptor Calculator")
        print("Choose the type of input:")
        print("1. SMILES string")
        print("2. MDL MolFile")
        print("3. SDF file")
        print("4. Command Line Wrapper")
        print("5. Exit")

        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == '1':
            get_descriptors_from_smiles()
        elif choice == '2':
            get_descriptors_from_mdl()
        elif choice == '3':
            get_descriptors_from_sdf()
        elif choice == '4':
            command_line_wrapper()
        elif choice == '5':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
