```python

import os
from google.colab import files

# Download the GNINA binary
!wget https://github.com/gnina/gnina/releases/download/v1.1/gnina
!chmod +x gnina

# Upload the PDB file
uploaded = files.upload()
for filename in uploaded.keys():
    with open('rec.pdb', 'wb') as f:
        f.write(uploaded[filename])

# Upload the SDF file
uploaded = files.upload()
for filename in uploaded.keys():
    with open('lig.sdf', 'wb') as f:
        f.write(uploaded[filename])

# Redock the cognate ligand
!./gnina -r rec.pdb -l lig.sdf --autobox_ligand lig.sdf -o docked.sdf --seed 0

# Parse the GNINA output
with open('docked.sdf', 'r') as f:
    lines = f.readlines()
docking_score = float(lines[1].split()[1])
rmsd = float(lines[2].split()[1])
binding_affinity = float(lines[3].split()[1])
pose_rank = int(lines[4].split()[1])
ligand_efficiency = float(lines[5].split()[1])
intermolecular_energy = float(lines[6].split()[1])
internal_energy = float(lines[7].split()[1])
torsion_energy = float(lines[8].split()[1])
electrostatic_energy = float(lines[9].split()[1])
van_der_waals_energy = float(lines[10].split()[1])

# Display the results
print("Docking Score:", docking_score, "kcal/mol")
print("RMSD:", rmsd, "Å")
print("Binding Affinity:", binding_affinity, "μM")
print("Pose Rank:", pose_rank)
print("Ligand Efficiency:", ligand_efficiency)
print("Intermolecular Energy:", intermolecular_energy, "kcal/mol")
print("Internal Energy:", internal_energy, "kcal/mol")
print("Torsion Energy:", torsion_energy, "kcal/mol")
print("Electrostatic Energy:", electrostatic_energy, "kcal/mol")
print("Van der Waals Energy:", van_der_waals_energy, "kcal/mol")
```
